```python
##3	We have BRICS countries in top 20 country list, create a bar chart representing their renewable energy consumption.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
countries=pd.read_csv(r"C:\Users\ANJUM AHMED\Downloads\top20CountriesPowerGeneration.csv")
countries

```

```python
brics=countries[(countries.Country=="Brazil")|(countries.Country=="Russia")|(countries.Country=="India")|(countries.Country=="Canada")]
brics
```

```python
brics1=countries.loc[[2,3,4,6],["Country","Total "]]
brics1
```

```python
Total_Consumption=brics.plot.bar(x='Country',y='Total ',color="purple")
Total_Consumption

plt.xlabel("Country")
plt.ylabel("Total")
plt.legend(["Country"])
plt.title('Consumption of BRICS')
```

```python
##9	Create a pie chart which shows the percentage of energy consumption in 2020 on international group level(BRICS,OECD,CIS)
```
